# PyCity Schools Analysis

Spend Per Capita:
 - Interestingly, the more dollars that are budgeted per student, the worse the learning outcomes are. Schools with less than $585 performed the best, with the highest overall passing rates of 90%. 
 - As the spend-per-student ranges increase, the overall passing rates drop by approx. 9% for each spend bucket. The highest spending range ($645-$680/student) has the lowest overall passing rates of 54%.
 - While it's unclear from the data about what is driving this trend, this indicates that higher spend per student does not automatically result in better student performance. 

School Size:
 - Schools with student populations less than 2,000 appear to perform significantly better than schools between 2,000 and 5,000, with 32% more students passing overall. This data cut suggests that it's more helpful for students' performance to have more schools with smaller populations. 
 - It's possible that schools with > 2,000 students having lower passing rates is a result of a higher student-to-teacher ratio, but we do not have the data on number of teachers per school, so it's unclear if we can truly draw this conclusion.

School Type:
 - Charter schools appear to perform better than district (public) schools, with approx. 37% more students passing overall. 

Math and Reading Scores by Grade:
 - The math and reading scores across grades 9 through 12 is relatively stable, with only a few percentage point changes from freshman to senior year. This data suggests that the average scores in math and reading are less dependent on the year in school and more dependent on the school itself.

Highest vs. Lowest Performing Schools:
 - The main takeways from this data cut is that all of the top performing schools are charter schools, compared to the lowest performing schools that are district (public) schools. This was shown in the School Type data cut, but further proven by this representation.
 - Furthermore, the average  per-student-budget for the top 5 highest performing schools is an average of $606, compared to the lowest schools' average per-student-budget of $647. This trend was discussed more in the Spend Per Capita analysis, and this data cut further suggests that there is an inverse correlation to per-student-budget and overall passing rates.

```python
# Dependencies and setup
import pandas as pd
from pathlib import Path
```

```python
# Load resource files
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read school and student data files and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
# Calculate the total budget
school_budget = school_data_complete["budget"].unique()

total_budget = school_budget.sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_count
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_count
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[    
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame(
    {"Total Number of Unique Schools": [school_count],
     "Total Students": [student_count],
     "Total Budget": [total_budget],
     "Average Math Score": [average_math_score],
     "Average Reading Score": [average_reading_score],
     "% Passing Math": [passing_math_percentage],
     "% Passing Reading": [passing_reading_percentage],
     "% Overall Passing": [overall_passing_rate]}
)

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}%".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}%".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:.2f}%".format)

# Display the DataFrame
district_summary
```

## Scores by School Spending

```python
# Establish spending bins 
spending_bins = [0, 585, 630, 645, 680]
spend_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the merge_school_df since values are unformatted (unlike per_school_summary)
school_spending_df = merge_school_df.copy()

# Drop same columns as in per_school_summary
school_spending_df = school_spending_df.drop(dropped_columns,axis=1)

# Rename column names as in per_school_summary
school_spending_df = school_spending_df.rename(columns={'school_name': 'School Name', 
                                                        'type': 'School Type',
                                                        'size': 'Total Students',
                                                        'budget': 'Total School Budget',
                                                        'per_school_capita': 'Per Student Budget',
                                                        'math_score': 'Average Math Score', 
                                                        'reading_score': 'Average Reading Score', 
                                                        'per_school_passing_math': '% Passing Math',
                                                        'per_school_passing_reading': '% Passing Reading', 
                                                        'overall_passing_rate': '% Overall Passing'
})

school_spending_df.head()
```

```python
# Use 'pd.cut' to categorize spending based on spending bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], 
                                                             spending_bins, 
                                                             labels=spend_labels)
school_spending_df.head()
```

```python
#  Calculate averages for the desired columns
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
          "Average Math Score" : spending_math_scores,
          "Average Reading Score": spending_reading_scores,
          "% Passing Math": spending_passing_math,
          "% Passing Reading": spending_passing_reading,
          "% Overall Passing": overall_passing_spending})

# Formatting
spending_summary["Average Math Score"] = spending_summary["Average Math Score"].map("{:.2f}%".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:.2f}%".format)
spending_summary["% Passing Math"] = spending_summary["% Passing Math"].map("{:.2f}%".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].map("{:.2f}%".format)
spending_summary["% Overall Passing"] = spending_summary["% Overall Passing"].map("{:.2f}%".format)

# Display results
spending_summary
```
